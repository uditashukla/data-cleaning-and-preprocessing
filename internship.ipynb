```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.head()
import pandas as pd

# STEP 1: Handle missing values
# Option 1: Remove rows with missing Income
df = df.dropna(subset=["Income"])

# STEP 2: Remove duplicate rows
df = df.drop_duplicates()

# STEP 3: Standardize column names (lowercase, underscores)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# STEP 4: Convert 'dt_customer' to datetime format
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y')

# STEP 5: Check and fix data types (if needed)
print(df.dtypes)

# STEP 6: Standardize categorical values (optional but useful)
df['education'] = df['education'].str.strip().str.title()
df['marital_status'] = df['marital_status'].str.strip().str.title()
df['marital_status'] = df['marital_status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single',
    'Divorced': 'Divorced',
    'Widow': 'Widowed',
    'Widower': 'Widowed',
    'Married': 'Married',
    'Together': 'Married',
})

# Normalize Education (map similar categories together)
df['education'] = df['education'].replace({
    'Basic': 'Basic',
    'Basic.4y': 'Basic',
    'Basic.6y': 'Basic',
    'Basic.9y': 'Basic',
    '2n Cycle': 'Graduation',
    'Master': 'Postgraduate',
    'Graduation': 'Graduate',
    'Phd': 'Postgraduate'
})

df = df.drop(columns=["z_costcontact", "z_revenue"])

# Remove outliers in Year_Birth (e.g., age < 18 or > 100)
current_year = pd.Timestamp.now().year
df['age'] = current_year - df['year_birth']
df = df[(df['age'] >= 18) & (df['age'] <= 100)]


# STEP 7: Optional - Save the cleaned file
df.to_csv("marketing_campaign_cleaned.csv", index=False)
# Summary of changes
print("Cleaned dataset shape:", df.shape)
print("Final cleaned shape:", df.shape)
print(" Unique Marital Status:", df['marital_status'].unique())
print(" Unique Education Levels:", df['education'].unique())

```
